## Skewness
**Definition**
- Skewness is a statistical measure that tells us whether the distribution of data is symmetric or tilted (skewed) towards one side.
- Skewness tells us whether the majority of values lie on the left side or right side of the distribution.
- It helps us understand the shape of the data distribution.

## Why is Skewness Important?

Skewness tells us

- Whether data is normally distributed
- Whether Mean is reliable
- Which Machine Learning algorithm should be used
- Whether transformation is needed
- Whether outliers are present
- Which statistical test should be applied

**Mathematical Formula**



Population Skewness

The population skewness is given by:

$$
\text{Skewness} =
\frac{\sum_{i=1}^{N}(x_i-\mu)^3}{N\sigma^3}
$$

Where:

- $x_i$ = Individual observation
- $\mu$ = Population mean
- $\sigma$ = Population standard deviation
- $N$ = Total number of observations


## Types of Skewness

### 1. Symmetrical Distribution (Zero Skewness)

- Skewness = 0
- Perfectly balanced
- Bell-shaped curve distribution
- No long tail
- **Mean = Median = Mode**

**examples**

- Heights
- IQ Scores
- Measurement Errors

```python
# Import required library
import matplotlib.pyplot as plt

import pandas as pd

# Sample data
marks = [10,20,30,40,50,60,70,80,90]

# Create dataframe
df = pd.DataFrame({"Marks": marks})

# Calculate skewness
sk = df["Marks"].skew()
print("Skewness:", sk)

# Interpretation
if sk > 0:
    print("Positively Skewed")
elif sk < 0:
    print("Negatively Skewed")
else:
    print("Symmetric Data")

# Create a figure with 1 row and 3 columns
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# -----------------------------
# Histogram
# -----------------------------
axes[0].hist(df["Marks"], bins=6, edgecolor="black")
axes[0].set_title("Histogram")
axes[0].set_xlabel("Marks")
axes[0].set_ylabel("Frequency")

# -----------------------------
# Density Plot
# -----------------------------
df["Marks"].plot(kind="density", ax=axes[1])
axes[1].set_title("Density Plot")
axes[1].set_xlabel("Marks")
axes[1].set_ylabel("Density")

# -----------------------------
# Box Plot
# -----------------------------
axes[2].boxplot(df["Marks"], vert=True)
axes[2].set_title("Box Plot")
axes[2].set_ylabel("Marks")

# Adjust spacing between plots
plt.tight_layout()

# Display all plots
plt.show()
```

### 2. Positive Skewness (Right Skewed)

- Most observations are small.
- A few observations are very large.
- Those large values create a long tail on the right side.
- **Mean > Median > Mode**

**examples**
- salary of employees in an organisationm

```python
# Import required library
import matplotlib.pyplot as plt

import pandas as pd

# Sample data
marks = [45, 48, 50, 52, 55, 58, 60, 62, 95]

# Create dataframe
df = pd.DataFrame({"Marks": marks})

# Calculate skewness
sk = df["Marks"].skew()
print("Skewness:", sk)

# Interpretation
if sk > 0:
    print("Positively Skewed")
elif sk < 0:
    print("Negatively Skewed")
else:
    print("Symmetric Data")

# Create a figure with 1 row and 3 columns
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# -----------------------------
# Histogram
# -----------------------------
axes[0].hist(df["Marks"], bins=6, edgecolor="black")
axes[0].set_title("Histogram")
axes[0].set_xlabel("Marks")
axes[0].set_ylabel("Frequency")

# -----------------------------
# Density Plot
# -----------------------------
df["Marks"].plot(kind="density", ax=axes[1])
axes[1].set_title("Density Plot")
axes[1].set_xlabel("Marks")
axes[1].set_ylabel("Density")

# -----------------------------
# Box Plot
# -----------------------------
axes[2].boxplot(df["Marks"], vert=True)
axes[2].set_title("Box Plot")
axes[2].set_ylabel("Marks")

# Adjust spacing between plots
plt.tight_layout()

# Display all plots
plt.show()
```

### 3. Negative Skewness (Left Skewed)

- Most observations are large.
- Few observations are very small.
- Long tail appears on the left.
- **Mean < Median < Mode**

```python
# Import required library
import matplotlib.pyplot as plt

import pandas as pd

# Sample data
marks = [20,90,91,92,93,94,95,96,97]

# Create dataframe
df = pd.DataFrame({"Marks": marks})

# Calculate skewness
sk = df["Marks"].skew()
print("Skewness:", sk)

# Interpretation
if sk > 0:
    print("Positively Skewed")
elif sk < 0:
    print("Negatively Skewed")
else:
    print("Symmetric Data")

# Create a figure with 1 row and 3 columns
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# -----------------------------
# Histogram
# -----------------------------
axes[0].hist(df["Marks"], bins=6, edgecolor="black")
axes[0].set_title("Histogram")
axes[0].set_xlabel("Marks")
axes[0].set_ylabel("Frequency")

# -----------------------------
# Density Plot
# -----------------------------
df["Marks"].plot(kind="density", ax=axes[1])
axes[1].set_title("Density Plot")
axes[1].set_xlabel("Marks")
axes[1].set_ylabel("Density")

# -----------------------------
# Box Plot
# -----------------------------
axes[2].boxplot(df["Marks"], vert=True)
axes[2].set_title("Box Plot")
axes[2].set_ylabel("Marks")

# Adjust spacing between plots
plt.tight_layout()

# Display all plots
plt.show()
```

## How to Reduce Skewness

If your data is highly skewed, you can apply transformations:

1. **Log Transformation** (np.log() or np.log1p()) – useful for right-skewed data such as income or sales.
2. **Square Root Transformation** (np.sqrt()) – suitable for moderately skewed count data.
3. **Cube Root Transformation** (np.cbrt()) – works for both positive and negative values.
4. **Box-Cox Transformation** (positive values only) – often used to make data more normal.
5. **Yeo-Johnson Transformation** – similar to Box-Cox but also supports zero and negative values.
6. **Outlier Treatment** (remove, cap, or winsorize extreme values) if the skewness is caused by a few extreme observations.
