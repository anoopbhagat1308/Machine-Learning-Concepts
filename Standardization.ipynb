## Standardization (Z-Score Scaling)

**Definition:**

+ Standardization is a feature scaling technique that transforms data so that each feature has a **mean of 0** and a **standard deviation of 1**. It preserves the shape of the original distribution while changing the scale of the values.

**formula**:

$$
Z = \frac{X - \mu}{\sigma}
$$

Where:

- x = original value
- μ = mean
- σ = standard deviation

**In simple words:**
- It tells us how far a value is from the mean in terms of standard deviations.

**Python Implementation**

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.DataFrame({
    'Age': [25, 35, 45],
    'Salary': [30000, 50000, 70000]
})

scaler = StandardScaler()

scaled_data = scaler.fit_transform(df)

print(scaled_data)
```

## When to Use Standardization?

Use Standardization when:

✅ Data is approximately normally distributed.

✅ Features have different scales.

✅ There are some outliers in the data.

✅ You are using algorithms that assume data is centered around the mean.

**Common Algorithms**
- Linear Regression
- Logistic Regression
- Support Vector Machine (SVM)
- Principal Component Analysis (PCA)
- Neural Networks
