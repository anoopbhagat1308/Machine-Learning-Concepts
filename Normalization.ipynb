## Normalization (Min-Max Scaling)

**Definition:**

+ Normalization is a feature scaling technique that rescales data to a fixed range, typically 0 to 1, without changing the relationships among the data points.

**Formula**:

$$
X_{scaled} = \frac{X - X_{min}}{X_{max} - X_{min}}
$$
	​

- In simple words:
It converts all values to a common scale where:

  + Minimum value becomes 0
  + Maximum value becomes 1
  + All other values fall between 0 and 1

**Python code implementation**

```python
import pandas as pd

# create a dummy data
df = pd.DataFrame({
    'Age': [25, 35, 45],
    'Salary': [30000, 50000, 70000]
})
```

```python
from sklearn.preprocessing import MinMaxScaler

# implementing MinMaxScaler
scaler = MinMaxScaler()

scaled_data = scaler.fit_transform(df)

print(scaled_data)
```

## When to Use Normalization?

Use Normalization when:

✅ You need all values within a fixed range (usually 0–1).

✅ The algorithm relies on distances or magnitudes.

✅ The data does not necessarily follow a normal distribution.

**Common Algorithms**
+ K-Nearest Neighbors (KNN)
+ K-Means Clustering
+ Deep Learning / Neural Networks
+ Image Processing
